## Feature Engineering 

### Handling Missing Values

```python
#Since the features ['Gender','Married','Self_Employed','Credit_History'] are missing at random
#So we are imputing each such feature with the mode of each corresponding column

def impute_nan(df,variable):
    most_frequent = df[variable].mode()[0]
    df[variable].fillna(most_frequent,inplace = True)
```

```python
for feat in ['Gender','Married','Self_Employed','Credit_History']:
    impute_nan(df,feat)
```

```python
df.isnull().sum()
```

```python
#Since Dependents has more frequent categories so we are reolacing the Nan values with
#some another category say 'Missing'

df['Dependents'] = np.where(df['Dependents'].isnull(),'Missing',df['Dependents'])
```

```python
df.isnull().sum()
```

#####  Here we are using Random Sample Imputation 

```python
df['Loan_Amount_Term_new'] = df['Loan_Amount_Term']
random_sample = df['Loan_Amount_Term'].dropna().sample(df['Loan_Amount_Term'].isnull().sum(),random_state=0)  # Taking random sample from Loan_Amount_Term
random_sample.index = df[df['Loan_Amount_Term'].isnull()].index   # Assigning indexes of null values to random samples,so that missing values can be replaced at correct index
df.loc[df['Loan_Amount_Term'].isnull(),'Loan_Amount_Term_new']=random_sample  # Assigning random samples to null values in Loan_Amount_Term

df['Loan_Amount_Term'] = df['Loan_Amount_Term_new']     #Assigning back to the original column name

df.drop(['Loan_Amount_Term_new'],axis = 1,inplace = True) 
```

```python
df.head()
```

```python
df.isnull().sum()
```

##### Using KNN Imputer to impute Nan Values in LoanAmount

```python
imputer = KNNImputer(n_neighbors = 100)
array = imputer.fit_transform(df[['LoanAmount']])

df1 = pd.DataFrame(array,columns = ['LoanAmount_new'])
df = pd.concat([df,df1],axis = 1)  # Concatenate new df1 created with the original df
df['LoanAmount'] = df['LoanAmount_new']    ##Assigning back to the original column name
df.drop(['LoanAmount_new'],axis = 1,inplace = True)  
```

```python
df.head()
```

### So we have handled the missing values

```python
df.isnull().sum()
```

### Now Let's handle Categorical features

##### Since  ['Married','Self_Employed','Loan_Status','Gender','Education','Property_Area'] are Nominal categorical data so
##### we are using one hot encoding

```python
def encode(df,variable):
    return pd.get_dummies(df[[variable]],drop_first = True)

for feat in ['Married','Self_Employed','Loan_Status','Gender','Education','Property_Area']:
    df = pd.concat([df,encode(df,feat)],axis = 1)
    
df.drop(['Married','Self_Employed','Loan_Status','Gender','Education','Property_Area'],axis = 1,inplace = True)
df.head()
```

##### Since Dependents_new is a ordinal categorical feature so we may use Label Encoding, here 
##### we are using Frequency Encoding

```python
df['Dependents'].unique()
```

```python
Encoded = df['Dependents'].value_counts().to_dict()  # Dictionary with frequency of each unique items
df['Dependents'] = df['Dependents'].map(Encoded)  # Mapping categorical value to their frequency
```

### Handling Imbalanced Dataset

```python
#Here number of 1's are more than double of the number of 0's
#So dataset is imbalanced

df['Loan_Status_Y'].value_counts()
```

```python
X = df.drop(['Loan_Status_Y'],axis = 1)  #Independent values
y = df['Loan_Status_Y']   #Dependent feature
```

##### We are using SMOTE  technique , SMOTETomek is a hybrid of upsampling and downsampling techniques of handling imbalanced dataset

```python
smk = SMOTETomek(sampling_strategy = .75)
X_res,y_res=smk.fit_sample(X,y)
X_res.shape,y_res.shape
```

```python
#So now our dataset is quite balanced, since there is no much difference between number of 0's and 1's

print(f'Original dataset shape {Counter(y)}')
print(f'Resampled dataset shape {Counter(y_res)}')
```

##### Also this algorithm is not sensitive to outliers

```python
X = X_res
y = y_res
```

##### Using RandomizedSearchCV and then feeding the best parameters to GridSearchCV
##### Since RandomizedSearchCV narrow down the result and then GridSearchCV would find the best parameters.
##### Those parameters would be the best for RandomForestClassifier.


```python

n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]

max_features = ['auto', 'sqrt','log2']

max_depth = [int(x) for x in np.linspace(10, 1000,10)]

min_samples_split = [2, 5, 10,14]

min_samples_leaf = [1, 2, 4,6,8]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion':['entropy','gini']}
print(random_grid)
```

```python
rf=RandomForestClassifier()
rf_randomcv=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_iter=10,cv=5,verbose=2,
                               random_state=100,n_jobs=-1)

rf_randomcv.fit(X_train,y_train)
```

```python
rf_randomcv.best_params_
```

```python
best_random_grid=rf_randomcv.best_estimator_
```

```python
param_grid = {
    'criterion': [rf_randomcv.best_params_['criterion']],
    'max_depth': [rf_randomcv.best_params_['max_depth']],
    'max_features': [rf_randomcv.best_params_['max_features']],
    'min_samples_leaf': [rf_randomcv.best_params_['min_samples_leaf'], 
                         rf_randomcv.best_params_['min_samples_leaf']+2],
    'min_samples_split': [rf_randomcv.best_params_['min_samples_split'], 
                          rf_randomcv.best_params_['min_samples_split'] +1,
                          rf_randomcv.best_params_['min_samples_split'] + 2],
    'n_estimators': [rf_randomcv.best_params_['n_estimators'], 
                     rf_randomcv.best_params_['n_estimators'] + 100]
}
```

## Automated Hyperparameter Tuning

#### Bayesian Optimization
Bayesian optimization uses probability to find the minimum of a function. The final aim is to find the input value to a function which can gives us the lowest possible output value.It usually performs better than random,grid and manual search providing better performance in the testing phase and reduced optimization time.
In Hyperopt, Bayesian Optimization can be implemented giving 3 three main parameters to the function fmin.

- Objective Function = defines the loss function to minimize.
- Domain Space = defines the range of input values to test (in Bayesian Optimization this space creates a probability distribution for each of the used Hyperparameters).
- Optimization Algorithm = defines the search algorithm to use to select the best input values to use in each new iteration.

```python
space = {'criterion': hp.choice('criterion', ['entropy', 'gini']),
        'max_depth': hp.quniform('max_depth', 10, 1200, 10),
        'max_features': hp.choice('max_features', ['auto', 'sqrt','log2', None]),
        'min_samples_leaf': hp.uniform('min_samples_leaf', 0, 0.5),
        'min_samples_split' : hp.uniform ('min_samples_split', 0, 1),
        'n_estimators' : hp.choice('n_estimators', [10, 50, 300, 750, 1200,1300,1500])
    }
```

```python
def objective(space):
    model = RandomForestClassifier(criterion = space['criterion'], max_depth = space['max_depth'],
                                 max_features = space['max_features'],
                                 min_samples_leaf = space['min_samples_leaf'],
                                 min_samples_split = space['min_samples_split'],
                                 n_estimators = space['n_estimators'], 
                                 ) 
    
    accuracy = cross_val_score(model, X_train, y_train, cv = 5).mean()

    # We aim to maximize accuracy, therefore we return it as a negative value
    return {'loss': -accuracy, 'status': STATUS_OK }
```

 ##### tpe = Tree-structured Parzen Estimator
 
##### ‘trials’, it is an object that contains or stores all the statistical and diagnostic information such as hyperparameter, loss-functions for 

##### each set of parameters that the model has been trained. ‘fmin’, it is an optimization function that minimizes the loss 

##### function and takes in 4 inputs. Algorithm used is ‘tpe.suggest’ , other algorithm that can be used is ‘tpe.rand.suggest’.

```python
trials = Trials()
best = fmin(fn= objective,
            space= space,
            algo= tpe.suggest,
            max_evals = 80,
            trials= trials)
best
```

```python
#Above hyperparameters have been obtained in the form of numbers , so we have to do mapping as below

crit = {0: 'entropy', 1: 'gini'}
feat = {0: 'auto', 1: 'sqrt', 2: 'log2', 3: None}
est = {0: 10, 1: 50, 2: 300, 3: 750, 4: 1200,5:1300,6:1500}


print(crit[best['criterion']])
print(feat[best['max_features']])
print(est[best['n_estimators']])
```

```python
trainedforest = RandomForestClassifier(criterion = crit[best['criterion']], max_depth = best['max_depth'], 
                                       max_features = feat[best['max_features']], 
                                       min_samples_leaf = best['min_samples_leaf'], 
                                       min_samples_split = best['min_samples_split'], 
                                       n_estimators = est[best['n_estimators']]).fit(X_train,y_train)
```

```python
predictionforest = trainedforest.predict(X_test)
print(confusion_matrix(y_test,predictionforest))
print('Accuracy :',accuracy_score(y_test,predictionforest))
print('-' * 75)
print(classification_report(y_test,predictionforest))
print('-' * 75)
acc5 = accuracy_score(y_test,predictionforest)
```
